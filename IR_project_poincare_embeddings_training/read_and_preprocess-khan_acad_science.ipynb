```python
from gzip import open as gopen

import pandas as pd
from pprint import pprint



def read_and_preprocess():
    science_data = pd.read_csv("train_khan_acad.csv")
    science_data_additional = pd.read_csv("val_khan_acad.csv")
    science_data_test = pd.read_csv("test_khan_acad.csv")

    combined = [science_data,science_data_additional,science_data_test]
    science_data = pd.concat(combined)
    taxonomy = science_data["hierarchy"]
    taxonomy_list = list(set(taxonomy.values.tolist()))
    print(len(taxonomy_list))
    cleaned_taxonomy = []
    for value in taxonomy_list:
        value = value.split(">>")
        value = [val.lower() for val in value]
        value="\t".join(value)
        cleaned_taxonomy.append(value)
    merged_syllabi = []
#     for syllabus in cleaned_taxonomy:
#       split_syllabus = syllabus.split("\t")
#       merged_syllabus = '\t'.join(split_syllabus[2:])
#       merged_syllabi.append(merged_syllabus)
    return cleaned_taxonomy








taxonomy = read_and_preprocess()
```

```python
taxonomy
```

```python

edges = []
nodes = set()
for index,tax in enumerate(taxonomy):
    phrases = tax.split("\t")
    for i,val in enumerate(phrases):
        if(i+1<len(phrases)):
            parent=val
            nodes.add(val)
            child = (phrases[i+1])
            nodes.add(child)
            edges.append( (child,parent) )
        else:
            nodes.add(val)
    if index<=10:print(tax.rstrip("\n"))
```

```python
print("Read {} nodes and {} edges".format( len(nodes), len(edges)) )
```

```python

with open("data/course_taxonomy_khan_acad.links.tsv","w") as fout:
    for src,dst in edges:
        fout.write("{}\t{}\n".format(src,dst))
    fout.close()
    
with open("data/course_taxonomy_khan_acad.nodes.txt","w") as fout:
    for node in nodes:
        fout.write("{}\n".format(node))
    fout.close()
```
