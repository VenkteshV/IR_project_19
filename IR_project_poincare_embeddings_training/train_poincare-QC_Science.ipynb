```python

import os
import logging
import numpy as np

from gensim.models.poincare import PoincareModel, PoincareKeyedVectors, PoincareRelations
from gensim.test.utils import datapath

logging.basicConfig(level=logging.INFO)
```

```python

file_path = 'data/course_taxonomy.links.tsv'

relations = PoincareRelations(file_path=file_path, delimiter='\t')
model = PoincareModel(train_data=relations)
```

```python
relations
```

```python
model = PoincareModel(train_data=relations, size=20, burn_in=10,epsilon=1e-06,negative=20) # train awith 10 burn-in epochs
model.train(epochs=1000, print_every=100)
```

```python
model.save("taxonomy_embedding_20.pkl")
```
